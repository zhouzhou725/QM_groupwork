```python
# Number of accidents by year
yearly_trend = data.groupby('accident_year').size().reset_index(name='Accident_Counts')
```

```python
# Plot the 2019-1023 trend change
# Draw a graph
plt.figure(figsize=(12, 8))
plt.plot(yearly_trend['accident_year'], yearly_trend['Accident_Counts'], marker='o', linestyle='-', color='b')
plt.title("Trends in the number of bicycle accidents from 2019 to 2023", fontsize=16, fontweight='bold')
plt.xlabel("Years", fontsize=14, fontweight='bold')
plt.ylabel("Accident_Counts", fontsize=14, fontweight='bold')
plt.grid(alpha=0.5)

# Save the drawing to a file
file_path = '/home/jovyan/work/Desktop/QM/group_work/Trends in the number of bicycle accidents from 2019 to 2023.png'
plt.savefig(file_path)
plt.show()
```

```python
# Visualize annual incident trends
plt.figure(figsize=(12, 8))
sns.lineplot(data=bicycle_trend, x='accident_year', y='total_accidents', marker='o', color='blue')
plt.title("Bicycle Accident Trends in London in 2023", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Total_accidents", fontsize=14, fontweight='bold')
plt.grid(True)


# save
file_path = '/home/jovyan/work/Desktop/QM/group_work/Bicycle Accident Trends in London in 2023.png'
plt.savefig(file_path)

plt.show()
```
